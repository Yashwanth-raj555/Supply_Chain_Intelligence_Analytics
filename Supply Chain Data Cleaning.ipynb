```python
# 1. Import Required Libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", "{:.2f}".format)

print("✅ Libraries imported successfully")
```

```python
# 2. Load Dataset
df = pd.read_csv("supply_chain.csv")

print(f"Dataset Shape: {df.shape[0]:,} rows × {df.shape[1]} columns")
df.head()
```

```python
# 3. Initial Exploration
# Dataset information
df.info()
```

```python
# Statistical summary for numeric + categorical columns
df.describe(include="all")
```

```python
# Missing values before cleaning
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({"Missing Count": missing, "Missing %": missing_pct})
missing_df[missing_df["Missing Count"] > 0]
missing_df.head()
```

```python
# Duplicate rows before cleaning
print("Duplicate Rows:", df.duplicated().sum())
```

```python
# Unique values in important categorical columns
cat_cols = ["Order Status", "Customer Segment", "Shipping Mode", "Delivery Status", "Market", "Department Name"]

for col in cat_cols:
    if col in df.columns:
        print(f"\n{col}:")
        print(df[col].unique())
```

```python
# 4. Standardize Column Names
print("Original Columns:")
print(df.columns.tolist())
```

```python
# Standardize column names
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(r"[()\s]+", "_", regex=True)
    .str.replace(r"[^a-z0-9_]", "", regex=True)
    .str.strip("_")
)
```

```python
# Rename selected columns for better readability
df.rename(columns={
    "order_date_dateorders": "order_date",
    "days_for_shipping_real": "days_shipping_real",
    "days_for_shipment_scheduled": "days_shipment_scheduled",
    "order_profit": "order_profit",
    "order_item_quantity": "order_quantity"
}, inplace=True)

print("✅ Standardized Columns:")
print(df.columns.tolist())
```

```python
# 5. Remove Redundant Columns
# Check if benefit_per_order and order_profit_per_order contain the same values

if {"benefit_per_order", "order_profit_per_order"}.issubset(df.columns):
    same_values = np.allclose(
        df["benefit_per_order"].fillna(0),
        df["order_profit_per_order"].fillna(0)
    )

    print("benefit_per_order same as order_profit_per_order:", same_values)

    if same_values:
        df.drop(columns=["benefit_per_order"], inplace=True)
        print("✅ Removed redundant column: benefit_per_order")

print("Current Shape:", df.shape)
```

```python
# 6. Handle Missing and Duplicate Values

# Remove fully empty rows
before_rows = df.shape[0]
df.dropna(how="all", inplace=True)
print(f"Removed fully empty rows: {before_rows - df.shape[0]}")
```

```python
# Fill numerical missing values with median
num_cols = df.select_dtypes(include=["int64", "float64"]).columns

for col in num_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].median())

# Fill categorical missing values with mode
cat_cols = df.select_dtypes(include=["object"]).columns

for col in cat_cols:
    if df[col].isnull().sum() > 0:
        mode_value = df[col].mode(dropna=True)
        df[col] = df[col].fillna(mode_value.iloc[0] if not mode_value.empty else "Unknown")

print("✅ Missing values handled")
print(df.isnull().sum())
```

```python
# Remove duplicate rows
before = df.shape[0]
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
after = df.shape[0]

print(f"✅ Removed {before - after} duplicate rows")
print("Current Shape:", df.shape)
```

```python
# 7. Clean Text and Categorical Columns

# Strip extra spaces from text columns
str_cols = df.select_dtypes(include="object").columns

for col in str_cols:
    df[col] = df[col].astype(str).str.strip()

print("✅ Extra spaces removed from text columns")
```

```python
# Standardize selected categorical columns to Title Case
cat_cols = [
    "order_status", "customer_segment", "shipping_mode", "delivery_status",
    "market", "department_name", "order_region", "order_country",
    "customer_state", "category_name"
]

for col in cat_cols:
    if col in df.columns:
        df[col] = df[col].str.title()

print("✅ Categorical text standardized")
```

```python
# 8. Date-Time Conversion

# Check raw date values
print(df["order_date"].head())
print("Current dtype:", df["order_date"].dtype)
```

```python
# Convert order_date to datetime
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")

print("New dtype:", df["order_date"].dtype)
print("Date Range:", df["order_date"].min(), "to", df["order_date"].max())
```

```python
# Extract useful date features
df["order_year"] = df["order_date"].dt.year
df["order_month"] = df["order_date"].dt.month
df["order_day"] = df["order_date"].dt.day
df["order_quarter"] = df["order_date"].dt.quarter
df["order_weekday"] = df["order_date"].dt.day_name()

df[["order_date", "order_year", "order_month", "order_day", "order_quarter", "order_weekday"]].head()
```

```python
# 9. Feature Engineering

# Shipping Delay = Actual shipping days - Scheduled shipping days
df["shipping_delay_days"] = df["days_shipping_real"] - df["days_shipment_scheduled"]

df["delivery_performance"] = np.where(
    df["shipping_delay_days"] > 0,
    "Delayed",
    np.where(df["shipping_delay_days"] < 0, "Early", "On Time")
)

print(df[["days_shipping_real", "days_shipment_scheduled", "shipping_delay_days", "delivery_performance"]].head())
```

```python
# Late Delivery Flag
df["is_late_delivery"] = (df["delivery_status"].str.lower() == "late delivery").astype(int)

print(df["is_late_delivery"].value_counts())
print(f"Late Delivery Rate: {df['is_late_delivery'].mean() * 100:.2f}%")
```

```python
# Profit Margin %
df["profit_margin_pct"] = np.where(
    df["sales"] != 0,
    (df["order_profit_per_order"] / df["sales"]) * 100,
    np.nan
).round(2)

df["profit_margin_pct"].describe()
```

```python
# Revenue Per Unit
df["revenue_per_unit"] = np.where(
    df["order_quantity"] != 0,
    df["sales"] / df["order_quantity"],
    np.nan
).round(2)

df["revenue_per_unit"].describe()
```

```python
# Order Size Category
df["order_size"] = pd.cut(
    df["order_quantity"],
    bins=[0, 1, 3, np.inf],
    labels=["Small", "Medium", "Large"],
    include_lowest=True
)

df["order_size"].value_counts()
```

```python
# Profit Type
df["profit_type"] = np.where(df["order_profit_per_order"] >= 0, "Profit", "Loss")

df["profit_type"].value_counts()
```

```python
# Sales Tier
df["sales_tier"] = pd.cut(
    df["sales"],
    bins=[-np.inf, 100, 300, np.inf],
    labels=["Low", "Medium", "High"]
)

df["sales_tier"].value_counts()
```

```python
# 10. Advanced Feature Engineering & Preprocessing

# 1. Country Name Standardization

country_map = {
    "Estados Unidos": "United States",
    "Alemania": "Germany",
    "Francia": "France",
    "Reino Unido": "United Kingdom",
    "Japón": "Japan",
    "Corea Del Sur": "South Korea",
    "Corea del Sur": "South Korea",
    "Australia": "Australia",
    "Canadá": "Canada",
    "Canada": "Canada",
    "Brasil": "Brazil",
    "México": "Mexico",
    "Italia": "Italy",
    "España": "Spain",
    "Países Bajos": "Netherlands",
    "China": "China",
    "India": "India"
}

if "order_country" in df.columns:
    df["order_country_en"] = df["order_country"].map(country_map).fillna(df["order_country"])
    print("Country name standardization completed.")
    display(df[["order_country", "order_country_en"]].head())
else:
    print("Column 'order_country' not found.")
```

```python
# 2. Season Feature Creation

if "order_month" in df.columns:
    df["order_season"] = df["order_month"].map({
        12: "Winter", 1: "Winter", 2: "Winter",
        3: "Spring", 4: "Spring", 5: "Spring",
        6: "Summer", 7: "Summer", 8: "Summer",
        9: "Fall", 10: "Fall", 11: "Fall"
    })
    print("Season feature created successfully.")
    display(df[["order_month", "order_season"]].head())
else:
    print("Column 'order_month' not found. Create order_month from order date first.")
```

```python
# 3. Outlier Flag Creation Instead of Removing Outliers

if "sales" in df.columns:
    Q1 = df["sales"].quantile(0.25)
    Q3 = df["sales"].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df["is_sales_outlier"] = ((df["sales"] < lower) | (df["sales"] > upper)).astype(int)

    print("Sales outlier flag created successfully.")
    print("Lower Limit:", lower)
    print("Upper Limit:", upper)
    print(df["is_sales_outlier"].value_counts())
else:
    print("Column 'sales' not found.")
```

```python
# 11. Negative Value and Outlier Detection

# Negative value checks
print("Negative Sales:", (df["sales"] < 0).sum())
print("Negative Order Profit:", (df["order_profit_per_order"] < 0).sum(), "(valid business losses, keep them)")
```

```python
# Outlier detection using IQR method
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns

outlier_summary = []

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    count = ((df[col] < lower) | (df[col] > upper)).sum()
    outlier_summary.append([col, count, lower, upper])

outlier_df = pd.DataFrame(outlier_summary, columns=["Column", "Outlier Count", "Lower Limit", "Upper Limit"])
outlier_df
```

```python
# 12. Final validation

print("=" * 60)
print("FINAL CLEANED DATASET SUMMARY")
print("=" * 60)
print(f"Total Rows      : {df.shape[0]:,}")
print(f"Total Columns   : {df.shape[1]}")
print(f"Missing Values  : {df.isnull().sum().sum()}")
print(f"Duplicate Rows  : {df.duplicated().sum()}")
print(f"Date Range      : {df['order_date'].min()} to {df['order_date'].max()}")
print("=" * 60)
```

```python
# Final data types
df.dtypes
```

```python
# Convert repeated categorical columns to category datatype
category_cols = [
    "order_status",
    "customer_segment",
    "category_name",
    "department_name",
    "shipping_mode",
    "delivery_status",
    "order_region",
    "market",
    "order_country",
    "customer_state",
    "delivery_performance",
    "order_weekday",
    "order_size",
    "profit_type",
    "sales_tier",
    "order_country_en",
    "order_season"
]

for col in category_cols:
    if col in df.columns:
        df[col] = df[col].astype("category")

print("✅ Categorical columns converted to category datatype")
```
